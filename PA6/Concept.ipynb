- Global Maximum : 극대화 문제의 최적화된 해, 검색 알고리즘을 통해 찾아내는 것이 목표
- Local Maximum : 주변 Neighbors들 사이에서 가장 높은 꼭대기

**Hill-Climing Search**는 현재 상황에서 더 최적화된 해가 있다면 이동하고, 더이상 이동이 불가하면 그곳을 Peak로 간주한다. 

Hill-Climing은 Local Maximum에서 해를 찾는 것을 멈춰 Local Optima에 빠질 수 있다. 그래서, 여러 번 무작위 초기 상태에서 알고리즘을 시작하는 Random Restart 전략을 사용한다.
